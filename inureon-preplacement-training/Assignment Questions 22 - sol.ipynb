ðŸ’¡ Question-1:

Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.

</aside>mm

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.val = value
        self.prev = None
        self.next = None

def binary_tree_to_dll(root):
    if not root:
        return None
    
    
    def inorder_helper(node, prev):
        nonlocal head
        
        if not node:
            return prev
        
        prev = inorder_helper(node.left, prev)
        
        
        dll_node = DoublyLinkedListNode(node.val)
        
        
        if prev:
            prev.next = dll_node
            dll_node.prev = prev
        else:
            
            head = dll_node
        
        
        prev = dll_node
        
        prev = inorder_helper(node.right, prev)
        
        return prev

    
    head = None
    inorder_helper(root, None)
    
    return head


def print_dll(head):
    if not head:
        return
    
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()


root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(5)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)


dll_head = binary_tree_to_dll(root)

print_dll(dll_head)
```

ðŸ’¡ Question-2

A Given a binary tree, the task is to flip the binary tree towards the right direction that is clockwise. See the below examples to see the transformation.

In the flip operation, the leftmost node becomes the root of the flipped tree and its parent becomes its right child and the right sibling becomes its left child and the same should be done for all left most nodes recursively

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def flip_binary_tree(root):
    if not root:
        return None
    
   
    if not root.left and not root.right:
        return root
    
   
    flipped_left = flip_binary_tree(root.left)
    flipped_right = flip_binary_tree(root.right)
    
    
    root.left = flipped_right
    root.right = flipped_left
    
    return root


def print_binary_tree(root):
    if not root:
        return
    
    queue = [root]
    while queue:
        node = queue.pop(0)
        print(node.val, end=" ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    print()

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)


print("Original Binary Tree:")
print_binary_tree(root)


flipped_root = flip_binary_tree(root)


print("Flipped Binary Tree:")
print_binary_tree(flipped_root)
```

ðŸ’¡ Question-3:

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def print_root_to_leaf_paths(root):
    if not root:
        return
    
    
    node_stack = [(root, str(root.val))]  
    path_stack = [str(root.val)]  
    
    while node_stack:
        node, path = node_stack.pop()
        current_path = path_stack.pop()
        
        
        if not node.left and not node.right:
            print("Path:", current_path)
        
        
        if node.left:
            node_stack.append((node.left, path + "->" + str(node.left.val)))
            path_stack.append(current_path + "->" + str(node.left.val))
        
        
        if node.right:
            node_stack.append((node.right, path + "->" + str(node.right.val)))
            path_stack.append(current_path + "->" + str(node.right.val))

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("Root-to-Leaf Paths:")
print_root_to_leaf_paths(root)
```

GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def build_tree_from_preorder_inorder(preorder, inorder):
    if not preorder or not inorder:
        return None

    root_val = preorder.pop(0)
    root_index = inorder.index(root_val)

    root = TreeNode(root_val)
    root.left = build_tree_from_preorder_inorder(preorder, inorder[:root_index])
    root.right = build_tree_from_preorder_inorder(preorder, inorder[root_index + 1:])

    return root

def build_tree_from_postorder_inorder(postorder, inorder):
    if not postorder or not inorder:
        return None

    root_val = postorder.pop()
    root_index = inorder.index(root_val)

    root = TreeNode(root_val)
    root.right = build_tree_from_postorder_inorder(postorder, inorder[root_index + 1:])
    root.left = build_tree_from_postorder_inorder(postorder, inorder[:root_index])

    return root

def are_traversals_same_tree(preorder, inorder, postorder):
   
    tree_from_pre_in = build_tree_from_preorder_inorder(preorder.copy(), inorder.copy())
    tree_from_post_in = build_tree_from_postorder_inorder(postorder.copy(), inorder.copy())

    
    return are_trees_same(tree_from_pre_in, tree_from_post_in)

def are_trees_same(tree1, tree2):
    if not tree1 and not tree2:
        return True
    if not tree1 or not tree2:
        return False

    return (
        tree1.val == tree2.val
        and are_trees_same(tree1.left, tree2.left)
        and are_trees_same(tree1.right, tree2.right)
    )


preorder = [1, 2, 4, 5, 3, 6, 7]
inorder = [4, 2, 5, 1, 6, 3, 7]
postorder = [4, 5, 2, 6, 7, 3, 1]

print(are_traversals_same_tree(preorder, inorder, postorder))  # Output: True
```
